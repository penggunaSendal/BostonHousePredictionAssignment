```python
# Checking outliers percentage

for k, v in data.items():
  q1 = v.quantile(0.25)
  q3 = v.quantile(0.75)
  irq = q3 - q1

  v_col = v[(v <= q1 - 1.5 * irq) | (v >= q3 + 1.5 * irq)]
  perc = np.shape(v_col)[0] * 100.0 / np.shape(data)[0]

  print('Column {} outliers = {:.2f}%'.format(k, perc))
```

```python
from scipy import stats
from scipy.stats import norm, skew

fig, axs = plt.subplots()
index = 0

for k, v in data.items():
  sns.distplot(data[k], fit=norm)

  (mu, sigma) = norm.fit(data[k])
  print('\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

  plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f})'.format(mu, sigma)], loc='best')

  fig = plt.figure()
  res = stats.probplot(data[k], plot=plt)
  plt.xlabel('{} Theoritical Quantiles'.format(k))
  plt.show()

  index += 1
```

```python
sns.distplot(data['MEDV'], fit=norm)

(mu, sigma) = norm.fit(data['MEDV'])
print('\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f})'.format(mu, sigma)], loc='best')
plt.ylabel('Frequency')
plt.xlabel('MEDV Dist.')

fig = plt.figure()
res = stats.probplot(data['MEDV'], plot = plt)
plt.show()
```

```python
# i try to remove skewness on MEDV

data['MEDV'] = np.log1p(data['MEDV'])

sns.distplot(data['MEDV'], fit=norm)

(mu, sigma) = norm.fit(data['MEDV'])
print('\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f})'.format(mu, sigma)], loc='best')
plt.ylabel('Frequency')
plt.xlabel('MEDV Dist.')

fig = plt.figure()
res = stats.probplot(data['MEDV'], plot = plt)
plt.show()
```

```python
plt.figure(figsize=(20, 10))

cor = data.corr().abs()
sns.heatmap(cor, annot=True)
plt.show()
```

```python
cor_target = abs(cor['MEDV'])
relevant_features = cor_target[cor_target > 0.2]
feature_names = [index for index, value in relevant_features.iteritems()]

feature_names.remove('MEDV')

print(feature_names)
print(len(feature_names))
```

```python
from sklearn import preprocessing

min_max_scaller = preprocessing.MinMaxScaler()
column_cels = feature_names

x = data.loc[:,column_cels]
y = data['MEDV']

x = pd.DataFrame(data=min_max_scaller.fit_transform(x), columns=column_cels)
fig, axs = plt.subplots(ncols=6, nrows=2, figsize=(20, 10))
index = 0
axs = axs.flatten()

for i, k in enumerate(column_cels):
  sns.regplot(y=y, x=x[k], ax=axs[i])

plt.tight_layout()
```

```python
# MODEL BUILDING
```

```python
# SPLITTING TEST AND TRAIN DATASET

from sklearn.model_selection import train_test_split

x = data[feature_names]
y = data['MEDV']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Linear Regression

from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(x_train, y_train)
```

```python
prediction = lr.predict(x_test)

print('Actual Value of The House = $ {}'.format(y_test[0]))
print('Model Prediction Value of The House = $ {}'.format(prediction[0]))
```

```python
from sklearn.metrics import mean_squared_error

score_map = {}

Linear_Regression_mse = mean_squared_error(y_test, prediction)
Linear_Regression_rmse = np.sqrt(Linear_Regression_mse)

score_map['LinearRegression'] = Linear_Regression_mse, Linear_Regression_rmse

print('MSE = {}'.format(Linear_Regression_mse))
print('RMSE = {}'.format(Linear_Regression_rmse))
```

```python
# Support Vector Regression

from sklearn.svm import SVR

svr = SVR()

svr.fit(x_train, y_train)
```

```python
svr_predic = svr.predict(x_test)

print('Actual Value of The House = $ {}'.format(y_test[0]))
print('Model Prediction Value of The House = $ {}'.format(svr_predic[0]))
```

```python
svr_mse = mean_squared_error(y_test, svr_predic)
svr_rmse = np.sqrt(svr_mse)

score_map['SVR'] = svr_mse, svr_rmse

print('MSE = {}'.format(svr_mse))
print('RMSE = {}'.format(svr_rmse))
```

```python
# KNeighborsRegression

from sklearn.neighbors import KNeighborsRegressor

knr = KNeighborsRegressor()

knr.fit(x_train, y_train)
```

```python
knr_prediction = knr.predict(x_test)

print('Actual Value of The House = $ {}'.format(y_test[0]))
print('Model Prediction Value of The House = $ {}'.format(knr_prediction[0]))
```

```python
knr_mse = mean_squared_error(y_test, knr_prediction)
knr_rmse = np.sqrt(knr_mse)

score_map['KNeighborsRegressor'] = knr_mse, knr_rmse

print('MSE = {}'.format(knr_mse))
print('RMSE = {}'.format(knr_rmse))
```

```python
# DEcision ttree regressor

from sklearn.tree import DecisionTreeRegressor

dtr = DecisionTreeRegressor(max_depth=5)

dtr.fit(x_train, y_train)
```

```python
dtr_predict = dtr.predict(x_test)


print('Actual Value of The House = $ {}'.format(y_test[0]))
print('Model Prediction Value of The House = $ {}'.format(dtr_predict[0]))
```

```python
dtr_mse = mean_squared_error(y_test, dtr_predict)
dtr_rmse = np.sqrt(dtr_mse)

score_map['DecisionTreeRegressor'] = dtr_mse, dtr_rmse

print('MSE = {}'.format(dtr_mse))
print('RMSE = {}'.format(dtr_rmse))
```

```python
# gradient boost

from sklearn.ensemble import GradientBoostingRegressor

gbr = GradientBoostingRegressor()

gbr.fit(x_train, y_train)
```

```python
gbr_predict = gbr.predict(x_test)


print('Actual Value of The House = $ {}'.format(y_test[0]))
print('Model Prediction Value of The House = $ {}'.format(gbr_predict[0]))
```

```python
gbr_mse = mean_squared_error(y_test, gbr_predict)
gbr_rmse = np.sqrt(gbr_mse)

score_map['GradientBoostingRegressor'] = gbr_mse, gbr_rmse

print('MSE = {}'.format(gbr_mse))
print('RMSE = {}'.format(gbr_rmse))
```

```python
score_ALL_map = pd.DataFrame(score_map, index=['MSE', 'RMSE'])

score_ALL_map
```

```python
plt.figure(figsize=(15, 10))

ax = sns.boxplot(data=score_ALL_map)

ax.set_title('Model Performance Comparison')
```
